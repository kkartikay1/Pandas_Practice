```python
import pandas as pd
```

```python
names = pd.Series(['Marks','Rita','Vicky','Justin','Jhosan','Michal'])
print(names)
```

```python
print(names.iloc[0:4])
```

```python
print(names.iloc[:5])
```

```python
print(names.iloc[:])
```

```python
print(names.iloc[::2])
```

```python
print(names)
```

```python
print(names.loc[::2])
```

```python
print(names.loc[:5])
```

```python
names = pd.Series(['Marks','Rita','Vicky','Justin','Jhosan','Michal'],index=[x for x in 'abcdef'])
print(names)
```

```python
print(names.iloc['a':'c'])
```

```python
print(names.loc['a':'c'])
```

```python
print(names)
```

```python
print(names.loc['c':])
```

```python
print(names.loc['c'::2])
```

```python
print(names.loc['a'::2])
```
