```python
import pandas as pd
```

```python
# List Of tuples Data
data1=[('Jack',34,'Sydney',5),("Rita",41,"Delhi",17),("Aadi",46,"New York",11)]
data2=[("Mothi",45,"Tokyo",11),("Veena",31,"Londan",10),("Shaun",36,"Las Vages",12)]
data3=[("Manik",47,"Sydney",14),("Vikas",42,"Delhi",11),("Samir",42,"Paris",15)]
# Create DataFrame Object for list
firstdf=pd.DataFrame(data1,columns=['Name','Age','City','Experience'],index=['a','b','c'])
seconddf=pd.DataFrame(data2,columns=['Name','Age','City','Experience'],index=['d','e','f'])
thirddf=pd.DataFrame(data3,columns=['Name','Age','City','Experience'],index=['g','h','i'])
print(firstdf)
print(seconddf)
print(thirddf)
```

```python
# Concatenate three DataFrame Along with Rows
df=pd.concat([firstdf,seconddf,thirddf],axis=0)
```

```python
print(df)
```

```python
# Concatenate three DataFrame Along with columns
df=pd.concat([firstdf,seconddf,thirddf],axis=1)
print(df)
```
