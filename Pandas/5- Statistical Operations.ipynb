```python
import pandas as pd
```

```python
import numpy as np
```

```python
s=pd.Series([10,20,13,4,0,12])
print("Given Series \n",s,type(s))
sm=s.sum()
print("Sum of Series of Elements=",sm)
```

```python
s=pd.Series([10,20,13,np.NaN,4,0,12,None])
print("Given Series \n",s,type(s))
sm=s.sum()
print("="*50)
print("Sum of Series of Elements=",sm)
```

```python
s=pd.Series([10,20,13,4,0,12,None])
print("Given Series \n",s,type(s))
sm=s.sum()
print("="*50)
print("Sum of Series of Elements=",sm)
```

```python
s=pd.Series([10,20,13,4,0,12])
print("Given Series \n",s,type(s))
sm=s.count()
print("Count of Series of Elements=",sm)
```

```python
s=pd.Series([10,20,13,4,0,12,None])
print("Given Series \n",s,type(s))
sm=s.count()
print("Count of Series of Elements=",sm)
print("Total No of element on series=",s.size)
```

```python
s=pd.Series([10,20,13,4,0,12,None])
print("Given Series \n",s,type(s))
print("="*50)
print("Max of Series of Elements=",s.max())
print("Min of Series of Elements=",s.min())
```

```python
s=pd.Series([10,20,13,4,0,12,None])
print("Given Series \n",s,type(s))
print("="*50)
print("Mean of Series of Elements=",s.mean())
```

```python
s=pd.Series([10,20,13,4,0,12,None])
print("Given Series \n",s,type(s))
print("="*50)
print("Round of Series of Elements=",s.round())
```

```python
s=pd.Series([10,20,13,4,0,12,None])
print("Given Series \n",s,type(s))
print("="*50)
print("Median of Series of Elements=",s.median())
```

```python
a=np.array([1,2,3,4])
s=pd.Series(a)
print(s)
print("="*50)
v=s.var()
print("Variance =",v)
```

```python
 a=np.array([1,2,3,4])
s=pd.Series(a)
print(s)
print("="*50)
v=s.std()
print("Standard Devitaion=",v)
```

```python
s=pd.Series([10,20,30,10,20,40,50,10,20,10])
print("Series Element \n",s)
m=s.mode()
print("mode =",m)
```
