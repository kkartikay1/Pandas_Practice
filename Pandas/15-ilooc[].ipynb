```python
import pandas as pd
import numpy as np
```

```python
# Get the data from Sutdent.csv into Datafram
df=pd.read_csv("D:\\Akash\\Python\\KVR-PYTHON-PART-1\\PANDAS\\NOTES\\peoples.csv")
print("="*50)
print("{}".format(df))
print("="*50)
```

```python
# DataFramwobj.iloc[row_number]
x=df.iloc[7]
print(x,type(x))
```

```python
x=df.iloc[6]
print(x,type(x))
```

```python
# DataFramwobj.iloc[row_number,col Number....]
x=df.iloc[7,1]
print(x,type(x))
```

```python
x=df.iloc[7,3]
print(x,type(x))
```

```python
# DataFramwobj.iloc[row_number,[col Number....]]
x=df.iloc[7,[1,2]]
print(x,type(x))
```

```python
x=df.iloc[6,[3,1]]
print(x,type(x))
```

```python
# DataFramwobj.iloc[Row Start:Row stop,[col Number....]]
x=df.iloc[3:9,[3,2]]
print(x,type(x))
```

```python
# DataFramwobj.iloc[Row Start:Row stop,col start:col stop]
# DataFramwobj.iloc[Row Start:Row stop:step,col start:col stop:step]
x=df.iloc[::2,::]
print(x,type(x))
```

```python
x=df.iloc[::2,:3]
print(x,type(x))
```

```python
x=df.iloc[::2,:3][::-1]
print(x,type(x))
```

```python
x=df.iloc[::2,::-1]
print(x,type(x))
```
