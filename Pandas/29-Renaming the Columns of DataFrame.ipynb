```python
import pandas as pd
# Crating DataFrame
df=pd.DataFrame({'A':[1,2,3],"B":[4,5,6],"C":[7,8,9]})
print("-------The DataFrame is--------")
print(df)
print("="*50)
# Renameing The Column
print(df.rename(columns={"A":'z',"B":'y',"C":'x'}))
```
