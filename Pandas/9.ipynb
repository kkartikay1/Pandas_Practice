```python
import pandas as pd
import numpy as np
```

```python
names=['Rossum','Ram','Travis','ANIL','kvr',np.nan,'Car',None,'dog','cat']
s=pd.Series(names)
print(s,type(s))
```

```python
# Print Lower Case element
lower=[str(name).lower() for name in s]
print("Lower Case Name \n",lower)
```

```python
# Print Lower Case element in Series() 
lower=pd.Series([str(name).lower() for name in s])
print("Lower Case Name \n",lower)
```

```python
# Print Upper Case element
lower=[str(name).upper() for name in s]
print("Upper Case Name \n",lower)
```

```python
# Print Upper Case element in Series() 
lower=pd.Series([str(name).upper() for name in s])
print("Upper Case Name  \n",lower)
```

```python
# Print Upper Case element in Series() 
lower=pd.Series([str(name).upper() for name in s]).sort_values()
print("Upper Case Name in Accending Order \n",lower)
```

```python
# Printthe Length of all element
nameswithlength={str(name).upper():len(str(name)) for name in names}
print("Length of all element Name \n",nameswithlength)
```

```python
print("Name\t\tLength")
for name,length in nameswithlength.items():
    print("{}\t\t{}".format(name,length ))
```
