```python
lst=[[11,"Jack",44,"Sydney",19],[12,"Rita",41,"Delhi",17],[13,"Aadi",46,"Mumbai",11],
    [14,"Mothi",45,"Delhi",15],[15,"Veena",43,"Delhi",14],[16,"Shaunak",42,"Mumbai",17],
    [17,"Manik",42,"Sydney",14],[18,"Vikas",42,"Delhi",11],[19,"Samir",42,"Mumbai",15],
    [20,"Shobit",40,"Sydney",12]]
df=pd.DataFrame(lst,columns=["CID","NAME","AGE","CITY",'EXPR'])
print(df)
```

```python
# Get the first group
df.groupby("CITY").first()
```

```python
# Get the first group Name
df.groupby("CITY").first()[["NAME"]]
```

```python
# Get the first group Name and AGE
df.groupby("CITY").first()[["NAME","AGE"]]
```

```python
# Get the city group size
df.groupby("CITY").size()
```

```python
# Get the mean of age group 
values=df.groupby("CITY").agg({'AGE':'mean',"EXPR":"mean"})
print(values)
```

```python
# Get the mean and sumof age group 
values=df.groupby("CITY").agg({'AGE':'mean',"EXPR":"sum"})
print(values)
```

```python
# Get the mean of age group 
values=df.groupby("CITY")["AGE"].agg({'size','mean',"sum"})
print(values)
```

```python
# Get the mean of age group 
values=df.groupby("CITY")["AGE"].agg({'size','mean',"sum",'median',"std"})
print(values)
```

```python
# Get the mean of age group 
values=df.groupby("CITY")["AGE"].agg({'size','mean',"sum",'median',"std",'var'})
print(values)
```
