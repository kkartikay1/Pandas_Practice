```python
import numpy as np
a=np.array((100,200,300,400,500))
print(a,type(a))
s=pd.Series(a,dtype=float)
print(s,type(s))
```

```python
a=np.array((100,200,300,400,500))
print(a,type(a))
print("="*50)
s=pd.Series(a,index=["ID"+str(int(x)) for x in range(1,len(a)+1)],dtype=float)
print(s,type(s))
```
