```python
import pandas as pd
import numpy as np
```

```python
# dropna
```

```python
s=pd.Series([10,20,13,4,0,12,None])
print("Given Series \n",s,type(s))
print("="*50)
print("Median of Series of Elements=",s.dropna(inplace=True))
```

```python
# values_counts()
```

```python
s=pd.Series([10,20,30,10,20,40,20,50,10,20,10])
print("Series Element \n",s)
c=s.value_counts()
print("="*50)
print("Cound Valuees =\n",c)
```

```python
# Unique
```

```python
s=pd.Series([10,20,30,10,20,40,20,50,10,20,10])
print("Series Element \n",s)
c=s.unique()
print("="*50)
print("Unique Valuees =\n",c)
```

```python
s=pd.Series([10,20,30,10,20,40,20,50,10,20,10])
print("Series Element \n",s)
c=s.nunique()
print("="*50)
print("Number of Unique Valuees =\n",c)
```

```python
s=pd.Series([28,12,34,11,29,3],index=["Red","Blue","Wite","Green","Yellow","Black"])
print("Series Element \n",s)
```

```python
s=pd.Series([28,12,34,11,29,3],index=["Red","Blue","Wite","Green","Yellow","Black"])
print("Series Element \n",s)
sor=s.sort_values()
print("="*50)
print("Shorted Valuees =\n",sor)
```

```python
s=pd.Series([28,12,34,11,29,3],index=["Red","Blue","Wite","Green","Yellow","Black"])
print("Series Element \n",s)
sor=s.sort_values()[::-1]
print("="*50)
print("Shorted Valuees in Decendiong oreder=\n",sor)
```

```python
s=pd.Series([28,12,34,11,29,3],index=["Red","Blue","Wite","Green","Yellow","Black"])
print("Series Element \n",s)
sor=s.sort_values(ascending=False)
print("="*50)
print("Shorted Valuees in Decendiong oreder=\n",sor)
```

```python
s=pd.Series([28,12,34,11,29,3])
print("Series Element \n",s)
print("="*50)
print(s.idxmax())
```

```python
s=pd.Series([10,20,30,10,20,40,20,50,10,20,10])
print("Series Element \n",s)
print("="*50)
print("Maximum Element Index",s.idxmax())
```

```python
# pow() Power 
```

```python
s1=pd.Series([2,3,1,3])
s2=pd.Series([10,20,3,40])
s3=s2.pow(s1)
print("s1\n",s1)
print("s2\n",s2)
print("Power Series\n",s3)
```

```python
s1=pd.Series([2,3,1,3])
s2=pd.Series([10,20,3,40])
s3=s1.pow(s2)
print("s1\n",s1)
print("s2\n",s2)
print("Power Series\n",s3)
```

```python
# eq () Equale
```

```python
s1=pd.Series([2,3,1,3])
s2=pd.Series([10,20,3,40])
s3=s1.eq(s2)
print("s1\n",s1)
print("s2\n",s2)
print("Power Series\n",s3)
```

```python
s1=pd.Series([2,3,1,3])
s2=pd.Series([10,3,1,40])
s3=s1.eq(s2)
print("s1\n",s1)
print("s2\n",s2)
print("Power Series\n",s3)
```

```python
# gt ()
```

```python
s1=pd.Series([2,3,1,3])
s2=pd.Series([3,2,3,1])
s3=s1.gt(s2)
print("s1\n",s1)
print("s2\n",s2)
print("Power Series\n",s3)
```

```python
# lt()less thas eual to
```

```python
s1=pd.Series([2,3,1,3])
s2=pd.Series([10,3,1,40])
s3=s1.lt(s2)
print("s1\n",s1)
print("s2\n",s2)
print("Power Series\n",s3)
```

```python
s1=pd.Series([2,3,1,3])
s2=pd.Series([3,2,3,1])
s3=s1.eq(s2)
print("s1\n",s1)
print("s2\n",s2)
print("Power Series\n",s3)
```
