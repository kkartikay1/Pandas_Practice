```python
import pandas as pd
```

```python
names = pd.Series(['Marks','Rita','Vicky','Justin','Jhosan','Michal'])
print(names)
```

```python
names = pd.Series(['Marks','Rita','Vicky','Justin','Jhosan','Michal'],
                 index=['a','b','c','d','e','f'])
print(names)
```

```python
import numpy as np
```

```python
values = np.arange(100,700,100)
seriesObj=pd.Series(values,index=np.arange(3,21,3))
print(seriesObj)
```

```python
values = np.arange(100,700,100)
seriesObj=pd.Series(values,index=np.arange(8,14),dtype=float)
print(seriesObj)
```

```python
values = np.arange(100,700,100)
seriesObj=pd.Series(values,index=[x for x in 'abcdef'])
print(seriesObj)
```

```python
values = np.arange(10,15)
seriesObj=pd.Series(data=values*5,index=values)
print(seriesObj)
```

```python
seriesObj=pd.Series([101,"Akash",99.65,5+4j])
print(seriesObj,type(seriesObj))
```
