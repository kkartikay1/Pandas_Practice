```python
import pandas as pd
```

```python
s=pd.Series(["Python","Java","DataScience","Python","Java","DataScience"])
print(s,type(s))
print(len(s))
```

```python
# Apply categroy in data types
s=pd.Series(["Python","Java","DataScience","Python","Java","DataScience"],dtype="category")
print(s,type(s))
print("Number of values in Series object=",len(s))
```

```python
cat=pd.Categorical(['a','b','c','d','a','c','b','a','d'])
print(cat,type(cat))
```

```python
cat=pd.Categorical(['a','b','c','d','a','c','b','a','d'],['c','b','a'])
print(cat,type(cat))
```

```python
cat=pd.Categorical(['a','b','c','d','a','c','b','a','d'],['c','b','a'],ordered=True)
print(cat,type(cat))
```

```python
# creating Categorical
cat=pd.Categorical(['a','b','a','c',None,'b'],categories=['b','a','c'])
print(cat)
```

```python
cat=pd.Categorical(['a','c','c',None],categories=['b','a','c'])
df=pd.DataFrame({"cat":cat,"s":["a","c",'c',None]})
print("DataFarme Data")
print(df)
print("="*50)
print("Describe")
print(df.describe())
print("="*50)
print("A column of DataFrame -'Cat'- Series Data")
print(df["cat"].describe())
```

```python
cat=pd.Categorical(['a','c','c',None],categories=['b','a','c'])
print(cat.categories)
```

```python
s=pd.Series(['a','b','c',"a"],dtype='category')
print(s.cat.categories)
```

```python
for g in s.cat.categories:
    print(g)
```

```python
x=["Group %s"%g for g in s.cat.categories]
print(x)
```

```python
x=["Group %s"%g.upper() for g in s.cat.categories]
print(x)
```

```python
for val in ["Group %s"%g.upper() for g in s.cat.categories]:
    print(val)
```

```python
s=pd.Series(['a','b','c',"a"],dtype='category')
print(s.cat.categories)
x=["Group %s"%g.upper() for g in s.cat.categories]
print(x)
```

```python
# Adding Categories.add.categories() method for append new categoires
s=pd.Series(['a','b','c',"a"],dtype='category')
s=s.cat.add_categories([4])
print(s.cat.categories)
```
