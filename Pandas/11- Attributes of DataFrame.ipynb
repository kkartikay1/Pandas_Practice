```python
#Implement V=values attribbut in DataFrame object
# Creartig a 2D dictionary values as dictionary onject
import numpy as np
dict={"Sales":{"Name":'Rossum',"Age":67,"Gender":'Male'},
      "Marketing":{"Name":'Rossy',"Age":6,"Gender":'Female'}}
# Creating a data Frame Object
data_frame=pd.DataFrame(dict)
print("="*50)
print("Orignal DataFrame")
print("="*50)
print(data_frame)
print("="*50)
print("Values Attribute NumPy Array of this DataFrame is :")
print("="*50)
# Implementing values attribute on this dataframe
print(data_frame.values)
print("="*50)
print("Dimension of Data values in ndarray-numpy",np.ndim(data_frame.values))
print("Shape of Data values in ndarray-numpy",np.shape(data_frame.values))
```
